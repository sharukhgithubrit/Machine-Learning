```python
iris= pd.read_csv('/kaggle/input/iris/Iris.csv')
```

```python
iris.info()
```

```python
X = iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
X.head()
```

```python
y = iris['Species']

y.head()
```

```python
from sklearn.preprocessing import LabelEncoder 

le=LabelEncoder()

y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import AdaBoostClassifier

# Create an AdaBoostClassifier instance with specified parameters
abc = AdaBoostClassifier(n_estimators=50, learning_rate=1, random_state=0)

# Train the AdaBoost Classifier on the training data
model1 = abc.fit(X_train, y_train)

# Predict the response (class labels) for the test dataset
y_pred = model1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(' AdaBoost Classifier Model Accuracy:',accuracy_score(y_test,y_pred))
```

```python
# load required classifier
from sklearn.ensemble import AdaBoostClassifier

# import Support Vector Classifier
from sklearn.svm import SVC

# import scikit-learn metrics module for accuracy calculation
from sklearn.metrics import accuracy_score

# Create an SVC instance with probability=True and kernel='linear'
svc = SVC(probability=True, kernel='linear')

# Create an AdaBoostClassifier object with SVC as the base estimator
abc = AdaBoostClassifier(n_estimators=50, base_estimator=svc, learning_rate=1, random_state=0)

# Train the AdaBoost Classifier on the training data
model2 = abc.fit(X_train, y_train)

# Predict the response for the test dataset
y_pred = model2.predict(X_test)

# Calculate and print model accuracy
print("Model Accuracy with SVC Base Estimator:", accuracy_score(y_test, y_pred))
```
