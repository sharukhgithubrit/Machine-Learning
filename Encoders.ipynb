##  Binary encoding

```python
from category_encoders import BinaryEncoder
be = BinaryEncoder() #The purpose of using the BinaryEncoder is to encode categorical features into binary representations.
be.fit_transform(df['fuelType'])
```

## Label encoding

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance
le = LabelEncoder() #It's suitable for ordinal data with a clear order.t may not be appropriate for nominal data

# Fit and transform the 'fuelType' column
df['fuelType_encoded'] = le.fit_transform(df['fuelType'])
```

```python
df
```

## Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

# Create an OrdinalEncoder instance
oe = OrdinalEncoder() #ordinal and label are similar [Label encoding doesn't consider the order];

# Fit and transform the 'fuelType' column
df['fuelType_encoded'] = oe.fit_transform(df[['fuelType']])
```

```python
df
```

## Use ordinal encoding when your data has a clear order, and use label encoding when you need a simple numeric representation for nominal data without implying an order.
